```python
# 2.9 Iterables
# Show an example of every iterable in Table 2.5

# INSERT YOUR CODE HERE
# List (ordered, mutable, allows duplicates)
my_list = [1, 2, 3, 4]
for item in my_list:
    print(item)

# Tuple (ordered, immutable, allows duplicates)
my_tuple = (1, 2, 3, 4)
for item in my_tuple:
    print(item)

# String (ordered, immutable)
my_string = "Hello, World!"
for char in my_string:
    print(char)

# Set (unordered, mutable, does not allow duplicates)
my_set = {1, 2, 3, 4}
for item in my_set:
    print(item)

# Dictionary (unordered, mutable, key-value pairs)
my_dict = {'a': 1, 'b': 2, 'c': 3}
for key, value in my_dict.items():
    print(key, value)

# Range (immutable sequence of numbers)
my_range = range(5)
for num in my_range:
    print(num)

# Bytes (immutable sequence of bytes)
my_bytes = b'Hello'
for byte in my_bytes:
    print(byte)

# Bytearray (mutable sequence of bytes)
my_bytearray = bytearray(b'Hello')
for byte in my_bytearray:
    print(byte)
```

```python
# 2.12 Operations on Sets
# rand1 and rand2 are two independent lists of 1000 numbers between 1 and 1000 inclusive
# What are set1 - set4?
# set1 and set2 are sets created from rand1 and rand2, respectively. set3 is the union of set1 and set2, and set4 is the intersection of set1 and set2.
# How many numbers between 1 and 1000 appear in neither list?
import random
random.seed(42)

rand1 = [random.randint(1,1000) for i in range(1000)]
rand2 = [random.randint(1,1000) for i in range(1000)]
set1 = set(rand1)
set2 = set(rand2)
set3 = set1 | set2
set4 = set1 & set2
print(len(set1), len(set2), len(set3), len(set4))

# Calculate the numbers that appear in neither list
numbers_not_appearing = set(range(1, 1001)) - set3

# Print the count of numbers that appear in neither list
print(len(numbers_not_appearing))

```
