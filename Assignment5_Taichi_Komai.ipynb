```python
def factorial(n):
  if not isinstance(n, int) or n < 0:
     return "Factorial is defined only for positive integers. "

  numbers = [x for x in range(1, n + 1)]
  product = 1
  for x in numbers:
    product *= x
  return product

print(factorial(5))
print(factorial(10))
print(factorial(20))
print(factorial(50))
```

```python
 def special_factorial(n):
  if n < 0:
    return "Factorial is not defined for negative numbers."
  elif n == 0:
    return 1
  else:
    num_list = [x for x in range(1, n + 1)]

    if 4 in num_list:
      num_list.remove(4)
    if 13 in num_list:
      num_list.remove(13)
    if 22 in num_list:
      num_list.remove(22)

    product = 1
    for num in num_list:
      product *= num

    return product


print(special_factorial(5))
print(special_factorial(10))
print(special_factorial(20))
print(special_factorial(50))
```
